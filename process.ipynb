```python
from abc import ABC, abstractmethod

class IMusicProcess(ABC):
    @abstractmethod
    def process(): pass

class MP3Process(IMusicProcess):
    def process():
        pass
```

```python
DIR = r'C:\Users\user\Downloads\Music\hololive æœªåˆ†é¡æ­Œ - è¤‡è£½'
MUSIC_TYPE = ('.mp3',)
SAPERATERS = ('ï¼', '-', '_')
CLEAN_TEXT = (
    r'ã€.*?ã€‘',
    r'[(]?[Cc][Oo][Vv][Ee][Rr]([Ee][Dd] [Bb][Yy])?[)]?',
)
EXTRACT_NAME = (
    r'[ã€Œ](.*?)[ã€]',
    r'[ã€](.*?)[ã€]'
)
```

```python
def is_MP3(file):
    for type in MUSIC_TYPE:
        if file.suffix.lower() == type:
            return True
    return False

def clean_song_text(file_name):
    new_file_name = file_name
    for pattern in CLEAN_TEXT:
        regex = re.compile(pattern)
        new_file_name = regex.sub('', new_file_name).strip()
    
    for pattern in EXTRACT_NAME:
        regex = re.search(pattern, new_file_name)
        if regex is not None:
            indexes = regex.span()
            string_list = list(new_file_name)
            string_list[indexes[1] - 1] = SAPERATERS[0]
            del string_list[indexes[0]]
            new_file_name = ''.join(string_list)
    return new_file_name


def split_song_artist(file_name):
    song_name = ""
    artist = ""
    for saperator in SAPERATERS:
        try:
            index = file_name.index(saperator)
            song_name = file_name[:index].strip()
            artist = file_name[index + 1:].strip()
            break
        except:
            continue
    return song_name, artist
```

```python
for file in Path(DIR).iterdir():
    if file.is_file() and is_MP3(file):
        is_cover, text = (True, "cover ") if 'cover' in file.stem.lower() else (False, "")
        print(f'{text}Song: {file.stem}')
        file_name = clean_song_text(file.stem)
        print(f"after clean: {file_name}")

        song_name, artist = split_song_artist(file_name)
        print(f"after split: {song_name}{SAPERATERS[0]}{artist}")
        print('----------------------------')
    else:
        continue
```
